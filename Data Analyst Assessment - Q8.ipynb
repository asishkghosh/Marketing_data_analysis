```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objs as go
```

```python
df = pd.read_csv("data.csv")
#df = df.dropna()
df.head()
```

```python
# create a new dataframe with the necessary columns
df_channels = df[['AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5', 'AcceptedCmp1','AcceptedCmp2','Response', 'NumDealsPurchases', 'NumCatalogPurchases', 'NumWebPurchases', 'NumStorePurchases']]
channels = ['NumDealsPurchases', 'NumCatalogPurchases', 'NumWebPurchases', 'NumStorePurchases']
campaigns = ['AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5', 'AcceptedCmp1','AcceptedCmp2','Response'] 
```

```python
# calculate response rates for each channel
response_rates = []
for channel in channels:
    
    # select rows where at least one campaign was accepted
    accepted = df_channels[df_channels[campaigns].any(axis=1)]

    # calculate the mean response rate for the selected channel
    response_rate = accepted[channel].mean()
    
    response_rates.append(response_rate)
```

```python
# create a dataframe with the results
channels_df = pd.DataFrame({'Channel': ['Deals', 'Catalog', 'Web', 'Store'], 'Response Rate': response_rates})
channels_df
```
